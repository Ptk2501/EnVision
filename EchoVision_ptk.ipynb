```python
from PIL import Image
import matplotlib.pyplot as plt

catwall=Image.open("images/cat_on_wall.png")
beach=Image.open("images/beachcricket.png")

plt.figure(figsize=(6,4))
plt.imshow(catwall)
plt.axis('off')
plt.show()

plt.figure(figsize=(6,4))
plt.imshow(beach)
plt.axis('off')
plt.show()
```

```python
# object detection-HF
from transformers import pipeline

obj_det_model=pipeline(
    task="object-detection",
    model="facebook/detr-resnet-50"
)

beach_detections=obj_det_model(beach)
cat_detections=obj_det_model(catwall)

beach_detections
```

```python
def counts_to_natural_language(counts):
    """
    counts: Counter or dict -> {object_name: count}
    """
    if not counts:
        return "No objects were detected."

    phrases = []

    for obj, count in counts.items():
        if count == 1:
            phrases.append(f"a {obj}")
        else:
            phrases.append(f"{count} {obj}s")

    # Join phrases naturally
    if len(phrases) == 1:
        return f"The image contains {phrases[0]}."
    else:
        return "The image contains " + ", ".join(phrases[:-1]) + " and " + phrases[-1] + "."
```

```python
from collections import Counter

def count_objects(labels):
    """
    labels: list of object names (strings)
    """
    return Counter(labels)
```

```python
from collections import Counter

def count_objects_from_detections(detections):
    """
    detections: list of dicts with 'label' key
    """
    labels = [d["label"] for d in detections]
    return Counter(labels)
```

```python

beach_pic_desc=counts_to_natural_language(count_objects_from_detections(beach_detections))
print(beach_pic_desc)
cat_pic_desc=counts_to_natural_language(count_objects_from_detections(cat_detections))
print(cat_pic_desc)
```
