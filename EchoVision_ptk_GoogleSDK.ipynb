```python
from google import genai

f = open("keys/.gemini.txt")
key = f.read()

client = genai.Client(api_key=key)

print(key)
```

```python
from google import genai
from google.genai import types
custom_config = types.GenerateContentConfig(
    system_instruction="You are a helpful AI Assistant. Given an image perform object detection and provide a text output which contains the information about the labels detected and their counts.",
    temperature=1,
    top_p=0.8,
)

with open('images/cat_on_wall.png', 'rb') as f:
  image_bytes = f.read()

context = [
    types.Part.from_bytes(
        data=image_bytes,
        mime_type='image/png',
    ),
]

response = client.models.generate_content(
    model="gemini-3-flash-preview",
    contents=context,
)

print(response.text)
```

```python
from transformers import pipeline
import warnings
warnings.filterwarnings("ignore")
narrator = pipeline(task="text-to-speech", model="suno/bark-small")
```

```python
# Use a pipeline as a high-level helper
# Warning: Pipeline type "summarization" is no longer supported in transformers v5.
# You must load the model directly (see below) or downgrade to v4.x with:
# 'pip install "transformers<5.0.0'
from transformers import pipeline

summerizer = pipeline("summarization", model="facebook/bart-large-cnn")
```

```python
pic_summary=summerizer(response.text)[0]["summary_text"]
print(pic_summary)
```

```python
cat_pic_text=narrator(pic_summary)

```
